```python
import numpy as np
import pickle
import os
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from pyntcloud import PyntCloud
```

```python
dir = os.path.dirname(os.path.realpath('__file__'))
with open(dir + '/../pickles/35.pkl', 'rb') as f:
    filtered_lidar_pts = pickle.load(f, encoding='latin1')
```

```python
first_frame_lidar = filtered_lidar_pts[1]
```

```python
%matplotlib notebook

f2 = plt.figure()
ax2 = f2.add_subplot(111, projection='3d')

ax2.scatter(first_frame_lidar[:, 0],
            first_frame_lidar[:, 1],
            first_frame_lidar[:, 2],
            cmap='gray')

ax2.set_title('Lidar for cars')

plt.show()
```

```python
points = pd.DataFrame(first_frame_lidar, columns=['x', 'y', 'z'])
cloud = PyntCloud(points)
```

```python
cloud.plot()
```
