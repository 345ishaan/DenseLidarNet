```python
hf = h5py.File('../ground_truth/0035_lidar.h5', 'r')
first_frame_lidar = np.array(hf.get('lidar_0035_0_0'))
```

```python
first_frame_lidar
```

```python
%matplotlib notebook

f2 = plt.figure()
ax2 = f2.add_subplot(111, projection='3d')

ax2.scatter(first_frame_lidar[:, 0],
            first_frame_lidar[:, 1],
            first_frame_lidar[:, 2],
            cmap='gray')

ax2.set_title('Lidar for cars')

plt.show()
```
